### Data source

### Data characterisitcs

Publisher: Agricultural Research Service (Department of Agriculture)

The dataset includes different types of vegetation and their characteristics (such as stem length, stem diam and branches) in southern Califiornia. In this area stem weevils (M. janthiniformis) were released and used as a biological control agent of dalmatian toadflax (Linaria dalmatica). The data shows the spread of weevil populations as well as the impact on the growth of the different vegetation types.

## Analysis Traffic Crashes Chicago

### Data source

https://data.cityofchicago.org/Transportation/Traffic-Crashes-Crashes/85ca-t3if


### Data characterisitcs

Crash data shows information about each traffic crash on city streets within the City of Chicago limits and under the jurisdiction of Chicago Police Department (CPD).

Updated: November 11, 2023
Data Provided by: City of Chicago

Size of the dataset fullfills the reqirements of the project. Data is on daily basis containing 49 columns with 779483 rows. 

Multiple response variables available to choose from.

### Research question

#### Linear Regression

**Which aspects can help predict the number of units involved in a crash?**
- Y = number of units involved
- Column "NUM_UNITS", a unit can be a motor vehicle, a pedestrian, a bicyclist, or another non-passenger roadway user
- Possible predictor variables: Posted speed limit, traffic control device, weather condition, lighting condition, roadway surface condition, street alignment, work zone, injuries total, crash hour, crash day of week, crash month, crash location 

**Which aspects affect the number of injuries in a crash?**
- Y = Injuries total
- Column "INJURIES_TOTAL", total persons sustaining fatal, incapacitating, non-incapacitating, and possible injuries as determined by the reporting officer
- Possible predictor variables: Posted speed limit, traffic control device, weather condition, lighting condition, roadway surface condition, street alignment, work zone, injuries total, crash hour, number of units, road defect, trafficway type

#### Classification 

**Severity of the crash: Which aspects can help predict if an accident will result in injuries and/or require towing or if it will result in no injuries and the vehicles can drive away?**
- Column "CRASH_TYPE" can be either "Injury and/or Tow Due to Crash" or "No Injury / Drive Away"
- Possible predictor variables: Posted speed limit, weather condition, lighting condition, roadway surface condition, street alignment, road defect, work zone, trafficway type, damage

**Hit and Run: Which aspects can help predict whether a crash involved a hit and run, where a driver caused the crash and fled the scene without exchanging information and/or rendering aid?**
- Column "HIT_AND_RUN_I" can be either yes or no and defines if the crash did/did not involve a driver who caused the crash and fled the scene without exchanging information and/or rendering aid
- Possible predictor variables: Posted speed limit, weather condition, lighting condition, roadway surface condition, street alignment, work zone, number of units involved, injuries total, crash hour, traffic control device
